### Working with iris data

```python
from sklearn.datasets import load_iris
```

```python
import pandas as pd
```

```python
iris=load_iris()
# iris
iris.data
```

```python
iris.feature_names
```

```python
iris.target
```

```python
iris.target_names
```

```python
iris_data=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_data
```

```python
iris_data.isna().sum()
```

```python
iris_data['species']=iris.target
iris_data['species_name']=iris_data.species.apply(lambda x:iris.target_names[x])
iris_data
```

```python
iris_data=iris_data.drop('species',axis=1)
```

```python
iris_data
```

```python
iris_data.describe(include='all')
```

```python
X=iris_data.drop('species_name',axis=1)
y=iris_data['species_name']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train
```

#### MLPClassifier

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp=MLPClassifier().fit(X_train,y_train)
```

```python
y_pred=mlp.predict(X_test)
```

```python
mlp.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
