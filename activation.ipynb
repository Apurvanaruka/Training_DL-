## Activation Function and its types


```python
# import Libraries
import numpy as np
import matplotlib.pyplot as plt
```

### Linear Activation function

```python
def linear(x):
    
    return x

xdata = np.linspace(-6, 6, 100)
ydata = linear(xdata)
plt.plot(xdata, ydata)
# linear Simple Formula
# y = x
# it is make a simple straight line
plt.title('Linear Activation Function')
plt.xlabel('X_Data')
plt.ylabel('Y-Data')
plt.show()

```

### Binary Activation function


```python
def binary(x):
    if (x<=0):
        return 0
    else:
        return 1
    
data = binary(.2)
print(data)
# Binary Simple Formula
# 0 if x <= 0
# 1 if x > 0
# it is lie b/w [0,1]
```

### Non Linear Activation Function

### Types
#### (A) Sigmoid

```python
# Sigmoid/Logistic Activation function
def sigmoid(x):
    return 1/(1+np.exp(-x))

xdata = np.linspace(-10, 10, 50)
ydata = sigmoid(xdata)
plt.plot(xdata, ydata)
# Sigmoid Simple Formula
# 1 / (1 + e^(-x))
# it is lie b/w [0,1]
plt.title('Sigmoid Activation Function')
plt.xlabel('X_Data')
plt.ylabel('Y-Data')
plt.show()
```

#### (B) Tanh

```python
import numpy as np
import matplotlib.pyplot as plt

# Tanh Activation Function
def tanh(x):
    return np.tanh(x)

xdata = [i for i in range(-5,6)]
ydata = tanh(xdata)
plt.plot(xdata, ydata)
# tanh simple formula

# tanh(x) = 2/(1+e^(-2x)) - 1   or  (e^x -e^-x)/ (e^x +e^-x)
# it is lie b/w [-1,1]
plt.title('Tanh Activation Function')
plt.xlabel('X_Axes')
plt.ylabel('Y-Axes')
plt.show()
```

#### (C) ReLU (Reactified Linear Unit)

```python
# Relu Activation Function
def relu(x):
    data=[max(0,i)for i in x]
    r=np.array(data,dtype=float)
    return r

xdata = np.linspace(-10, 5, 50)
ydata = relu(xdata)
plt.plot(xdata, ydata)
# Relu Simple Formula
# Relu(x) = max(0,x)
# it is lie b/w [0,infinite(n values)]
plt.title('Relu Activation Function')
plt.xlabel('X_Axes')
plt.ylabel('Y_Axes')
plt.show()
```
