```python
import cv2
import signal

eid = 101
name = "Bhanu"
dept = "Account"
faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
clf = cv2.face.LBPHFaceRecognizer()
video_cap = cv2.VideoCapture(0)
video_cap.set(3, 340)  # width
video_cap.set(4, 480)  # height

def recognize(img, faceCascade, clf):
    try:
        cord = drawBoundary(img, faceCascade, 1.1, 10, (255, 25, 255), clf)
    except Exception as e:
        print("Error:", str(e))
    return img

def drawBoundary(img, classifier, scaleFactor, minNbr, color, clf):
    try:
        gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        features = classifier.detectMultiScale(gray_image, scaleFactor, minNbr)
        for (x, y, w, h) in features:
            cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 3)
            eid, predict = clf.predict(gray_image[y:y+h, x:x+w])
            confidence = int((100*(1-predict/300)))
            if confidence > 77:
                if eid == 101:
                    cv2.putText(img, f"Employe_ID:{eid}", (x, y-5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)
                    cv2.putText(img, f"Name:{name}", (x, y-5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)
                    cv2.putText(img, f"Dept:{dept}", (x, y-30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)
                else:
                    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 3)
                    cv2.putText(img, f"Unknown", (x, y-5), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)
    except Exception as e:
        print("Error:", str(e))
    return img

def handle_interrupt(signal, frame):
    video_cap.release()
    cv2.destroyWindow()
    exit(0)

# Register interrupt signal handler
signal.signal(signal.SIGINT, handle_interrupt)

while True:
    ret, img = video_cap.read()
    img = cv2.flip(img, 1)
    img = recognize(img, faceCascade, clf)
    cv2.imshow('video', img)
    k = cv2.waitKey(30)

    if k == 27:
        break

video_cap.release()
cv2.destroyWindow()
```
